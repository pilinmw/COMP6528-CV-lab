```python
def plot_histogram(img: np.ndarray) -> None:
    red_channel = img[:, :, 0].flatten()
    green_channel = img[:, :, 1].flatten()
    blue_channel = img[:, :, 2].flatten()
    _, axes = plt.subplots(1, 3, figsize=(12, 4))
    axes[0].hist(red_channel, bins=256, color="red", alpha=0.6)
    axes[0].set_title("Red")
    axes[1].hist(green_channel, bins=256, color="green", alpha=0.6)
    axes[1].set_title("Green")
    axes[2].hist(blue_channel, bins=256, color="blue", alpha=0.6)
    axes[2].set_title("Blue")
    plt.show()


plot_histogram(img)
```

```python
img_eq = histogram_equalization(img)
_, axes = plt.subplots(1, 2)
axes[0].imshow(img)
axes[0].set_title("Original")
axes[1].imshow(img_eq)
axes[1].set_title("Equalized")
```

```python
plot_histogram(img_eq)
```
